```python
# Q1

import tkinter as tk

win = tk.Tk()
win.geometry("300x200")


regL = tk.Label(win,text="regno")
regL.grid(row=0,column=0)
regIN = tk.Entry(win)
regIN.grid(row=0,column=1)


nameL = tk.Label(win,text="Name")
nameL.grid(row=1,column=0)
nameIN = tk.Entry(win)
nameIN.grid(row=1,column=1)


deptL = tk.Label(win,text="Dept")
deptL.grid(row=2,column=0)
deptIN = tk.Entry(win)
deptIN.grid(row=2,column=1)

gen = tk.IntVar()
genderL = tk.Label(win,text="gender")
genderL.grid(row=3,column=0)
male = tk.Radiobutton(win,text="male",value="male",variable=gen)
male.grid(row=3,column=1)
female = tk.Radiobutton(win,text="female",value="female",variable=gen)
female.grid(row=3,column=2)


ageL = tk.Label(win,text="Age")
ageL.grid(row=4,column=0)
ageIN = tk.Spinbox(win,from_=0,to=30)
ageIN.grid(row=4,column=1)

insert=tk.Button(win,text="insert")
insert.grid(row=5,column=1,pady=10)

update=tk.Button(win,text="update")
update.grid(row=5,column=2,pady=10)

delete=tk.Button(win,text="delete")
delete.grid(row=6,column=1,pady=10)

select=tk.Button(win,text="select")
select.grid(row=6,column=2,pady=10)

win.mainloop()
```
